Let $\mathcal{P}$ be a polygon with vertices $(x_1,y_1), (x_2,y_2), \ldots, (x_n,y_n)$, with $n \geq 3$ labelled counter-clockwise.

Recall that *Green's Theorem* states that for a positively oriented, piecewise continuous, simple closed curve $\partial D$ in the plane, we have:

$$ \oint_{\partial D} (P dx + Q dy) = \int \int_{D} \frac{\partial Q}{\partial x} - \frac{\partial P}{\partial y} dA  $$

If it happens that $\frac{\partial Q}{\partial x} - \frac{\partial P}{\partial y} = 1$, then $\int \int_{D} \frac{\partial Q}{\partial x} - \frac{\partial P}{\partial y} dA$ is equal to the area of the region $D$.

One way to achieve  $\frac{\partial Q}{\partial x} - \frac{\partial P}{\partial y} = 1$ is to set $P(x,y) = 0$ and $Q(x,y)=x$.

Since $\mathcal{P}$ is a simple closed curve, we must now find a positively oriented, piecewise continuous parametrization of $\mathcal{P}$.

From this point on, we assume, for simplicity of notation, that $n+1 = 0$, such that $(x_{n+1},y_{n+1}) = (x_1,y_1)$.

To do that, for each $1 \leq i \leq n$, we parametrize the line segment from $(x_i,y_i)$ to $(x_{i+1},y_{i+1})$ by:

$$ r_{i}(t) = \big( x_{i} + (x_{i+1}-x_{i})t, y_{i} + (y_{i+1}-y_{i})t \big) \; , \; 0 \leq t \leq 1 $$

Note that $r_{i}^{\prime}(t) = \big((x_{i+1}-x_{i}), (y_{i+1}-y_{i})\big)$.

Since $\partial \mathcal{P} = \bigcup_{i=1}^{n} r_{i}$ and by applying *Green's Theorem*, we have that:

$$ \int \int_{\mathcal{P}} \frac{\partial Q}{\partial x} - \frac{\partial P}{\partial y} dA = \oint_{\partial \mathcal{P}} (P dx + Q dy) = \sum_{i=1}^{n} \oint_{r_{i}} (P dx + Q dy)$$

Substituting $P(x,y) = 0$ and $Q(x,y) = x$ we get 

$$ \int \int_{\mathcal{P}} 1 dA = \sum_{i=1}^{n} \oint_{r_{i}}  Q dy = \sum_{i=1}^{n} \int_{0}^{1} \big[(x_{i+1}-x_{i})t + x_i\big] (y_{i+1}-y_{i}) dt =  \sum_{i=1}^{n} (x_{i+1}-x_{i})(y_{i+1}-y_{i}) \cdot \big[\frac{t^2}{2}\big]_{0}^{1} + x_{i}(y_{i+1}-y_{i}) \cdot \big[ t \big]_{0}^{1} = \sum_{i=1}^{n} \frac{1}{2}(x_{i+1} + x_{i})(y_{i+1} - y_{i}) = \sum_{i=1}^{n} \frac{-1}{2}(x_{i+1}y_{i} - x_{i}y_{i+1}) + \frac{1}{2}(x_{i+1}y_{i+1}) - \frac{1}{2}(x_{i}y{i}) = \sum_{i=1}^{n} \frac{-1}{2}(x_{i+1}y_{i} - x_{i}y_{i+1}) + \frac{1}{2} \sum_{i=1}^{n} x_{i+1}y_{i+1} - \frac{1}{2} \sum_{i=1}^{n} x_{i}y_{i} = \frac{-1}{2} \sum_{i=1}^{n} (x_{i+1}y_{i} - x_{i}y_{i+1}) $$

Let $V$ be the volume of $\mathcal{P}$.

Since $V = \int \int_{\mathcal{P}} 1 dA$, then $V = \frac{-1}{2} \sum_{i=1}^{n} (x_{i+1}y_{i} - x_{i}y_{i+1})$.

For polygons with vertices labelled clockwise, by a similar argument, we have $V = \frac{1}{2} \sum_{i=1}^{n} (x_{i+1}y_{i} - x_{i}y_{i+1})$.

So, in general, for any polygon, $V = \frac{1}{2} \left\vert  \sum_{i=1}^{n} (x_{i+1}y_{i} - x_{i}y_{i+1}) \right\vert $.

Suppose that the result is true for $n+1$.

Consider a polygon $\mathcal{P}$ with $n+2$ vertices.

Let $C$ be the number of triangulations of $\mathcal{P}$.

Take any two adjacent vertices $u,v$ of $\mathcal{P}$.

Let $F$ be the *contraction mapping* of the edge $\overline{uv}$, i.e, the mapping that maps all points in the segment $\overline{uv}$ to the same point $w$.

This mapping is continuous, such that any edge or diagonal incident to $u$ or $v$ are, after application of the *contraction mapping*, incident to the point $w$.

(I do not prove the existence and properties of this contraction mapping, but they are known results that can be found in many books about metric spaces.)

By applying $F$ to $\mathcal{P}$ we obtain a new polygon, call it $\mathcal{P}^{\prime}$ with $n+1$ vertices.

Let $T$ be a triangulation of $\mathcal{P}$.

Consider any vertex $z \neq u,v$.

Since $F$ is continuous, any line segment(and specially edges and diagonals) incident to $z$ is mapped to a line segment that is incident to $F(z)$.

Moreover, because $F$ is continuous, then any two non-intersecting line segments that are not incident to $u$ or $v$ are mapped by $F$ to two non-intersecting line segments that are not incident to $w$.

Also, since $F$ is continuous and $\mathcal{P}$ is convex, then $\mathcal{P}^{\prime}$ is also convex.

So, since $T$ is a triangulation of $\mathcal{P}$, then $F(T)$ is a triangulation of $\mathcal{P}^{\prime}$.

Now let $p$ be any point not colinear to $u$ and $v$ and consider the triangle $\triangle uvp$.

This triangle is mapped by $F$ to a single line segment.

So for any triangulation $T$ of $\mathcal{P}$, the triangle $\triangle uvp$ is mapped a single edge or diagonal of the triangulation $F(T)$ of $\mathcal{P}^{\prime}$.

So let $T^{\prime}$ be a triangulation of $\mathcal{P}^{\prime}$.

For each edge or diagonal incident to $w$ in $T^{\prime}$ we can expand(the reverse of contracting) that edge or diagonal to a triangle, resulting in a triangulation of $\mathcal{P}$.

Take a triangulation $T$ of $\mathcal{P}^{\prime}$ and let $e^{T}(w)$ be the degree(i.e number of incident edges and diagonals) of $w$ in this triangulation.

By the previous reasoning, we can obtain exactly $e$ triangulations of $\mathcal{P}$ by expanding edges and diagonals incident to $w$.

Let $\mathcal{T}$ be the set of triangulations of $\mathcal{P}^{\prime}$.

So $C = \sum_{T \in \mathcal{T}} e^{T}(w)$.

Since $\mathcal{P}$ is convex, we could use a similar argument for every other pair of adjacent vertices.

Since $\mathcal{P}$ has $(n+1)$ pair adjacent vertices in counter-clockwise order, then:

$$ (n+1) C = \sum_{z \in \mathcal{P}} \sum_{T \in \mathcal{T}} e^{T}(z) = \sum_{T \in \mathcal{T}} \sum_{z \in \mathcal{P}} e^{T}(z) = \sum_{T \in \mathcal{T}} 2(2n-1) = 2(2n-1) \sum_{T \in \mathcal{T}} 1 = 2(2n-1) \cdot C^{\prime} $$

where $C^{\prime} = \left\vert \mathcal{T} \right\vert$, i.e, the number of triangulations of $\mathcal{P}^{\prime}$.

Since $\mathcal{P}^{\prime}$ has $n+1$ vertices, then by the induction hypothesis $C^{\prime} = C_{n-1}$.

So 

$$ (n+1) C = 2(2n-1) \cdot C^{\prime} = 2(2n-1) \cdot \frac{1}{n} \cdot {2(n-1) \choose (n-1)} = 2(2n-1) \cdot \frac{1}{n} = \frac{(2n-2)!}{(n-1)!(n-1)!} = \frac{(2n-1)!}{n!} \cdot \frac{2}{(n-1)!} = \frac{(2n)!}{2n} \cdot \frac{1}{n!} \cdot \frac{2}{(n-1)!} = \frac{(2n)!}{n!n!} = {2n \choose n}  $$

and thus 

$$C = \frac{1}{n+1} {2n \choose n}$$

We start by proving the following result:

**Fact:** *For every $n \geq 3$, there is a polygon with $n$ vertices with exactly one triangulation.*

For the case $n = 4$, simply consider the polygon with vertices $(0,-1),(1,0),(0,1),(\frac{1}{2},0)$.

Now let $n \geq 4$.

We start by setting the vertices $v_0 = (0, 0), v_1 = (1,0), v_2 = (1,0)$.

There are still $n-3$ vertices left to be defined.

Let $m_1 = \big\lceil\frac{n-3}{2} \big\rceil$ and $m_2 = (n-2) - m_1$.

Note that $m_1 - 1 = m_2$.

Moreover, consider the line segment $\overline{v_{1}v_2}$.

For $1 \leq i \leq m_1$, define the point $u_{i}$ to be the point in which the line $x = \frac{i}{n}$ and the segment $\overline{v_{1}v_2}$ intersect.

Note that, by definition, all the points $u_i$, for $1 \leq i \leq m_1$ and the points $v_1$ and $v_2$ are colinear.

Finally, to define the remaining $m_2$ points, for each $1 \leq i \leq m_1 - 1 = m_2$, define the point $w_i$ to be the point of intersection between the line $y = 1$ and the line that passes through the points $v_0$ and $u_i$.

We have concluded definying all the vertices of the desired polygon $\mathcal{P}$.

Now we must show that it has a unique triangulation.

First, since ${v_0}{v_1}{v_2}$ is a triangle(and therefore convex) and, for all $1 \leq i \leq m_1$, $u_i \in \overline{v_{1}v_2}$, then each segment $\overline{{v_0}{u_i}} \subset \mathcal{P}$ and, thus, $\overline{{v_0}{u_i}}$ is a diagonal, for all $1 \leq i \leq m_1$.

This defines $m_1$ diagonals.

Moreover, for $1 \leq i \leq m_1 - 2 = m_2 - 1$, the segment $\overline{{u_i}{u_{i+1}}} \subset \overline{{v_1}{v_2}} \subset \mathcal{P}$.

Thus $\overline{{u_i}{u_{i+1}}}$ is a diagonal.

This defines another $m_2 - 1$ diagoanals.

So we have a set with $m_1 + (m_2 - 1) = n - 3$ diagonals.

By **Theorem 1.8** every triangulation of a $n$-polygon has exactly $n-3$ diagonals.

So this set of diagonals is in fact a triangulation.

Since there are no other diagonals besides the ones in this set, this triangulation is unique.

The code below uses Python and *Processing.py* to draw the polygon as described in the proof. We also draw its unique triangulation.

```python
## This codes requires Processing.py

import math

def polygon(n):
    vertices = [[0.0, 0.0], [0.0, 1.0]]
    m_1 = int((n-3)/2)
    print(m_1)
    for i in range(n-3):
        if(i % 2 == 0):
            x = (i+1)*(1.0/n)
            y = (1-x)*math.tan(0.25*math.pi)
        else:
            t = 1.0/y
            x = x*t
            y = y*t
        vertices.append([x,y])
    vertices.append([1.0, 0.0])
    return vertices

M = 11

vertices = polygon(M)

my_width = 800
my_height = 800

translation = [40, 40]

def setup():
    size(my_width, my_height)
    background(180)
    
def draw():
    fill(255, 255, 255)
    stroke(200, 124, 45)
    beginShape()
    for i in range(M):
        vertex(0.3*my_width*vertices[i][0] + translation[0], 0.3*my_height*vertices[i][1] + translation[1])
    endShape(CLOSE)
    stroke(127)
    for i in range(M):
        if(i % 2 == 0):
            line(translation[0], translation[1], 0.3*my_width*vertices[i][0] + translation[0], 0.3*my_height*vertices[i][1] + translation[1])
            if(i + 2 < M):
                line(0.3*my_width*vertices[i][0] + translation[0], 0.3*my_height*vertices[i][1] + translation[1], 0.3*my_width*vertices[i+2][0] + translation[0],
           0.3*my_height*vertices[i+2][1] + translation[1])
```

Take any vertex $v$ that lies in the line $y = 1$.

This vertex is an ear vertex.

Take $\epsilon > 0$ sufficiently small and attach a square of size $\epsilon$ to $v$, overlapping one the square's edges with one of the edges of $\mathcal{P}_{n-3}$ that are incident to $v$, also overllaping one of the square's vertices with $v$.

This defines a new polygon $\mathcal{P}$ with $n$ vertices.

$\epsilon$ should be small enough to ensure that none of these attached vertices form a diagonal with any of the vertices of $\mathcal{P}_{n-3}$.
 
Since the square has two triangulations and $\mathcal{P}_{n-3}$ has a unique triangulation, and we attached the two in a way that creates no new diagonals, then the resulting polygon $\mathcal{P}$ has exactly two diagonals.
